```python
top_10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(20).index]
```

```python
top_10
```

```python
for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)
    
data[['X2']+top_10].head(40)
```

```python
def one_hot_top(df,variable,top_x_labels):
    
    
    for label in top_x_labels:
        
        df[variable+'_'+label]=np.where(data[variable]==label,1,0)

data=pd.read_csv(r'C:\Users\HP\Desktop\dataset\test.csv',usecols=['X1','X2','X3','X4','X5','X6'])


one_hot_top(data,'X2',top_10)

data.head()
```

```python
X1_top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(20).index]

X1_top_10

one_hot_top(data,'X1',X1_top_10)

data.head()
```
